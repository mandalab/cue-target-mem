```python
import pandas as pd
```

```python
cox_study = pd.read_excel("Cox Study Phase.xlsx")
cox_test = pd.read_excel("Cox Test Phase.xlsx")
```

```python
merged_df = pd.merge(cox_test, cox_study, on=['subject', 'stim.string.left', 'block', 'stim.num.left'], how='inner', left_index= False, right_index=False)
```

```python
merged_df
```

```python
merged_df['stim.string.right'] = merged_df['Target']
merged_df.drop(columns=['Target'], inplace=True)
```

```python
merged_df
```

```python
merged_df.columns
```

```python
merged_df.drop(columns=['Unnamed: 0', 'condition_y', 'phase_y', 'trial_y'], inplace=True)
```

```python
merged_df.rename(columns={'condition_x': 'condition', 'phase_x': 'phase', 'trial_x' : 'trial'}, inplace=True)
```

```python
merged_df
```

```python
merged_df['resp.type.rescore']
```

```python
merged_df['stim.num.right_x'] = merged_df['stim.num.right_y']
merged_df.rename(columns={'stim.num.right_x': 'stim.num.right'}, inplace=True)
```

```python
merged_df.drop(columns=['cv.left', 'cv.right', 'distractor.resp', 'freq.left', 'freq.right',
       'old.left', 'old.right', 'resp', 'rt', 'stim.distractor',
       'stim.num.right_y', 'studied'], inplace=True)
```
